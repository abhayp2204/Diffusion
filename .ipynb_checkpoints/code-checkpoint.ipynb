```python
# Imports
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definitions
NUM_LENGTH_UNITS = 25
TIME_LIMIT = 50
TIME_STEP = 2.0
ORIGIN = int(NUM_LENGTH_UNITS/2)
THRESHOLD = 0.1

# Diffusion equation parameters
D = 0.01
dt = 0.05
dx = 0.08

# Heatmap
heatmap1 = plt.get_cmap('hot')
heatmap2 = 'viridis'
```

```python
def decent_chance(x):
    return (x > THRESHOLD)

def at_boundary(x):
    return (x == 0 or x == (NUM_LENGTH_UNITS - 1))
```

```python
# Generate probability matrix
posX = np.linspace(-1, 1, NUM_LENGTH_UNITS)
posY = np.linspace(-1, 1, NUM_LENGTH_UNITS)
time = np.arange(0, TIME_LIMIT, TIME_STEP)

# Initialize P
shape = (len(time) + 1, NUM_LENGTH_UNITS)
P = np.zeros(shape)

# Since the particle starts at origin at time t=0, probability=1
P[0][ORIGIN] = 1

# Loop to generate probability matrix
for t in range(len(time)):
    for x in range(len(posX)):
        
        # Boundary condition
        if at_boundary(x):
            P[t+1][x] = 0
            continue
            
        # Diffusion equation
        P[t+1, x] = P[t, x] + D*dt/(dx**2)*(P[t, x+1] - 2*P[t, x] + P[t, x-1])
        
# Display probability matrix
# P
```

```python
# Graph: Probability of particle to be at x at time t vs x
legend = []

# Generate plots
step = 10
for t in range(0, len(P) - step, step):
    
    # Ignore graphs which are too spread to prevent crowding
#     if not decent_chance(P[t][mid]):
#         continue
        
    # Plot graph
    plt.plot(posX, P[t])
    legend.append(f"time = {time[t]}")

# Graph attributes
plt.title("Probability of particle to be at x at time t vs Position")
plt.legend(legend)
plt.xlabel("Position")
plt.ylabel("Probability")
plt.show()
```

```python
dy = dx

def plot(Dx, Dy, time):
    
    # Initialize P
    shape = (len(time)+1, NUM_LENGTH_UNITS, NUM_LENGTH_UNITS)
    P = np.zeros(shape)
    
    # Since the particle starts at origin at time t=0, probability=1
    P[0][ORIGIN][ORIGIN] = 1

    # Iterate over the time steps
    for t in range(len(time)):
        for x in range(len(posX)):
            for y in range(len(posY)):
                
                # Boundary condition
                if at_boundary(x):
                    P[t+1][x][y] = 0
                    continue
                if at_boundary(y):
                    P[t+1][y][x] = 0
                    continue
                    
                # 2D Diffusion equation
                P[t+1][y][x] = P[t][y][x]
                P[t+1][y][x] += Dx*dt/(dx**2)*(P[t][y][x+1] + P[t][y][x-1] - 2*P[t][y][x])
                P[t+1][y][x] += Dy*dt/(dy**2)*(P[t][y+1][x] + P[t][y-1][x] - 2*P[t][y][x])
                
    # Heatmap
    plt.imshow(P[len(time)], extent = [-1, 1, -1, 1], cmap = heatmap1)
    
    # Plot graph
    if Dx == Dy:
        plt.title(f"Dx = Dy and {len(time)} timesteps")
    elif Dx > Dy:
        plt.title(f"Dx > Dy and {len(time)} timesteps")
    else:
        plt.title(f"Dx < Dy and {len(time)} timestamps")
    plt.show()
```

```python
np.seterr('raise')
Dx = 0.01
Dy = 0.01

# Plot graph for various timesteps
plot(Dx, Dy, np.linspace(0, 10, 10))
plot(Dx, Dy, np.linspace(0, 10, 100))
plot(Dx, Dy, np.linspace(0, 10, 1000))
plot(Dx, Dy, np.linspace(0, 10, 10000))
```
