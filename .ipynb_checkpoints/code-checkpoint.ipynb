```python
# Imports
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-1, 1, 25)
t = np.arange(0, 25, 0.05)

D = 0.01
dt = 0.05
dx = 0.08

prob = []
prob.append([0]*len(x))
prob[0][12] = 1
print(t.size)

for i in range(t.size):
    prob.append([0]*len(x))
    
    for pos in range(len(x)):
        if pos == 0 or pos == 24:
            prob[i+1][pos] = 0
            continue
        prob[i+1][pos] = prob[i][pos] + D*dt/(dx**2) * (prob[i][pos+1] + prob[i][pos-1] - 2*prob[i][pos])
len(prob)
```

```python
pos = np.linspace(-1, 1, 25)
time = np.arange(0, 25, 0.05)

D = 0.01
dt = 0.05
dx = 0.08

shape = (len(time), 25)
P = np.zeros(shape)
P[0][12] = 1

for t in range(len(time)):
    for x in range(len(pos)):
        if x == 0 or x == 24:
            P[t+1][x] = 0
            continue
        P[t+1, x] = P[t, x] + D*dt/(dx**2)*(P[t, x+1] - 2*P[t, x] + P[t, x-1])
P
```

```python
leg = []
for i in range(0, len(prob), 10):
    dist = prob[i]
    if dist[12] > 0.1:
        plt.plot(x, dist)
        leg.append(f"t={t[i]}")
plt.legend(leg)
plt.xlabel("Position")
plt.ylabel("Probability")
plt.show()
```

```python
leg = []
for i in range(0, len(prob), 10):
    dist = P[0][i]
    if P[12][i] > 0.1:
        plt.plot(x, P[0][i])
#         leg.append(f"t={t[i]}")
plt.legend(leg)
plt.xlabel("Position")
plt.ylabel("Probability")
plt.show()
```
